```python
df = df.drop(columns=['DE_wind_offshore_capacity', "DE_wind_onshore_capacity","DE_wind_onshore_profile","DE_wind_offshore_profile","DE_wind_profile"])
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm")

# Title formatting
plt.title("Correlation Heatmap", fontsize=16)

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Save with high resolution and proper bounding
plt.savefig("correlation_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
columns = list(df.columns.values)
columns.remove("utc_timestamp")
columns
```

```python
plt.figure(figsize=(16,16))
for i, col in enumerate(columns):
    plt.subplot(4, 3, i+1)
    sns.histplot(df, x=col, kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
```

```python
plt.figure(figsize=(16,16))
for i, col in enumerate(columns):
    plt.subplot(4, 3, i+1)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```
